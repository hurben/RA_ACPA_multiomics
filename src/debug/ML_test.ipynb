```python
train_data = '/Users/m221138/RA_acpa_multiomics/analysis/machine_learning/5fold/enet_1condition_p1/fs_network_percentage_posVSneg/5p/1fold/multiplex.fs.train.tsv'
test_data = '/Users/m221138/RA_acpa_multiomics/analysis/machine_learning/5fold/enet_1condition_p1/fs_network_percentage_posVSneg/5p/1fold/multiplex.fs.test.tsv'
```

```python
train_df = pd.read_csv(train_data, sep='\t', index_col=0)
train_df = train_df.T
test_df = pd.read_csv(test_data, sep='\t', index_col=0)
test_df = test_df.T
```

```python
train_df.loc[train_df.acpa == 0, 'acpa'] = 'neg'
train_df.loc[train_df.acpa == 1, 'acpa'] = 'pos'
```

```python
test_df.loc[test_df.acpa == 0, 'acpa'] = 'neg'
test_df.loc[test_df.acpa == 1, 'acpa'] = 'pos'
```

```python
test_df
```

```python
specified_classifier = RandomForestClassifier()
specified_classifier = GradientBoostingClassifier()
clf = specified_classifier
```

```python


y_train = train_df.iloc[:,0]
X_train = train_df.iloc[:,1:]

y_test = test_df.iloc[:,0]
X_test = test_df.iloc[:,1:]
```

```python
y_train
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
clf.feature_importances_
```

```python
len(y_pred)
```

```python
positive_class = 'pos'
negative_class = 'neg'

tp_count = 0
fp_count = 0
tn_count = 0
fn_count = 0

observed_class_list = list(y_test)
predicted_class_list = list(y_pred)

for i in range(len(observed_class_list)):
    print (i)
    observed_value = observed_class_list[i]
    predicted_value = predicted_class_list[i]
    print (observed_value, predicted_value)
    
    if observed_value == positive_class:
        if predicted_value == positive_class: #Truely RA
            tp_count += 1
            print ('tp')
        if predicted_value == negative_class: #
            fn_count += 1
            print ('fn')

    if observed_value == negative_class:
        if predicted_value == negative_class:
            tn_count += 1
            print ('tn c1')
        if predicted_value == positive_class:
            fp_count += 1
            print ('fp c1')
```
