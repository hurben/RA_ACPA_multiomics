```python
import pandas as pd
import numpy as np
import math
```

```python
#Two files that contains sample information.
new_patient_info_file = '../../patient_info_data/additional_patient_info_Feb27/Rheum_Biobank_All_Visits.csv'
new_eac_patient_info_file = '../../patient_info_data/additional_patient_info_Feb27/EarlyArthritisCohort_DATA_2021-06-05_2351.csv'

sample_info_file = '../../patient_info_data/additional_patient_info_Feb27/sample_identification_Feb22.csv'
control_sample_info_file = '../../patient_info_data/Davis_105_mixed_sample_info.csv'
main_patient_info_file = '../../patient_info_data/patient_info.v2.tsv'
```

```python
#find sampleID = mayoID 
sample_info_dict = {}
sample_mayoinfo_dict = {}

sample_info_df = pd.read_csv(sample_info_file)
r, c = sample_info_df.shape

for i in range(r):
    mayoID = str(sample_info_df["MayoClinic_ID"][i])
    sampleID = str(sample_info_df["sample_ID"][i])
    sample_info_dict[mayoID] = sampleID
    sample_mayoinfo_dict[sampleID] = mayoID
```

```python
control_sample_info_df = pd.read_csv(control_sample_info_file)
r, c = control_sample_info_df.shape

for i in range(r):
    mayoID = str(control_sample_info_df["case_mayoclinic_id"][i])
    sampleID = str(control_sample_info_df["SAMPLEID"][i])
    if mayoID not in list(sample_info_dict.keys()):
        sample_info_dict[mayoID] = sampleID
    if sampleID not in list(sample_mayoinfo_dict.keys()):
        sample_mayoinfo_dict[sampleID] = mayoID
```

```python
#get patient ID and other valus; store them into dict
main_info_dict = {}

main_data_df = pd.read_csv(main_patient_info_file, sep='\t', index_col=0)
patientID_list = list(main_data_df.index.values)
r, c = main_data_df.shape

variable_list = ['acpa', 'sex','age','bmi','rf','smoking']
for i in range(r):
    patientID = str(patientID_list[i]) #this is sampleID

    for variable in variable_list:

        main_info_dict[patientID, variable] = list(main_data_df[variable])[i]  
```

```python
temp_patientID_list = []
for patientID in patientID_list:
    temp_patientID_list.append(str(patientID))
patientID_list = temp_patientID_list
```

```python
#find patient ID, match with current main_patient_ID_dict
#get crp, accp, esr, das28-crp, das28-esr, CDAI
new_data_df = pd.read_csv(new_patient_info_file, sep=',', index_col=0)
r, c = new_data_df.shape
patientID_list_tracker = []

count = 0 
for i in range(r):
    visit = new_data_df["Visit Description"][i]
    #convert mayoID to string 
    try: 
        mayoID = int(new_data_df["Mayo Clinic #"][i])
        mayoID = str(mayoID)
    except ValueError:
        patientID = 'NA'

    #check whether mayoID exists in sample info dict
    try: 
        patientID = sample_info_dict[mayoID]
    except KeyError:
        patientID = 'NA'

    if patientID != 'NA':
        if patientID in patientID_list:
            if visit == 'Visit 1 (Baseline)':
                count += 1
                
                patientID_list_tracker.append(patientID)
                sjc = new_data_df["Swollen Joint Count (range: 00 to 28)"][i]
                tjc = new_data_df["Tender Joint Count (range: 00 to 28)"][i]
                dag = new_data_df["Physician Global Assessment of Disease Activity (range: 000 to 100 mm)"][i]
                pag = new_data_df["Patient Global Assessment of Disease Activity (range: 000 to 100 mm)"][i]
                rf = new_data_df["Rheumatoid factor titer (IU/mL)"][i]
                acpa = new_data_df["Anti-CCP titer (U/mL)"][i]
                crp = new_data_df["C-reactive protein (most recent) (mg/L)"][i]
                esr = new_data_df["ESR (most recent) (mm/hour)"][i]
                
                try: cdai = (pag/10) + (dag/10) + int(sjc) + int(tjc)
                except ValueError: cdai = 'nan'
                    
                try: sdai = (pag/10) + (dag/10) + int(sjc) + int(tjc) + float(crp)
                except ValueError: sdai = 'nan'
                    
                try: das28esr = (0.56 * math.sqrt(int(tjc))) + (0.28 * math.sqrt(sjc)) + (0.7 * math.log(esr)) + (0.014 * pag)
                except ValueError: das28esr = 'nan'
                    
                try: das28crp = (0.56 * math.sqrt(int(tjc))) + (0.28 * math.sqrt(sjc)) + (0.36 * math.log(crp +1)) + (0.014 * pag) + 0.96
                except ValueError: das28crp = 'nan'   
                    
                if mayoID == '3595718':
                    print (sjc,tjc,dag,pag)

                main_info_dict[patientID,'rf_tit'] = rf
                main_info_dict[patientID,'acpa_tit'] = acpa
                main_info_dict[patientID,'crp'] = crp
                main_info_dict[patientID,'esr'] = esr
                main_info_dict[patientID,'cdai'] = cdai
                main_info_dict[patientID,'sdai'] = sdai
                main_info_dict[patientID,'das28esr'] = das28esr
                main_info_dict[patientID,'das28crp'] = das28crp
                main_info_dict[patientID,'sjc28_base'] = sjc
                main_info_dict[patientID,'tjc28_base'] = tjc
                
print (count)
```

```python
#do the same process with eac file 
new_eac_patient_df = pd.read_csv(new_eac_patient_info_file)
r, c = new_eac_patient_df.shape

count = 0
for i in range(r):
    try: 
        mayoID = int(new_eac_patient_df["mc_number"][i])
        mayoID = str(mayoID)
    except ValueError:
        patientID = 'NA'

    #check whether mayoID exists in sample info dict
    try: 
        patientID = sample_info_dict[mayoID]
    except KeyError:
        patientID = 'NA'
   
    if patientID != 'NA':
        if patientID in patientID_list:
            patientID_list_tracker.append(patientID)
            count += 1

            sjc = new_eac_patient_df["sjc28_base"][i]
            tjc = new_eac_patient_df["tjc28_base"][i]
            dag = new_eac_patient_df["base_pga"][i]
            pag = new_eac_patient_df["pga"][i]
            rf = 'nan'
            acpa = 'nan'
            crp = new_eac_patient_df["crp"][i]
            esr = new_eac_patient_df["esr"][i]

            try: cdai = (pag/10) + (dag/10) + int(sjc) + int(tjc)
            except ValueError: cdai = 'nan'

            try: sdai = (pag/10) + (dag/10) + int(sjc) + int(tjc) + float(crp)
            except ValueError: sdai = 'nan'

            try: das28esr = (0.56 * math.sqrt(int(tjc))) + (0.28 * math.sqrt(sjc)) + (0.7 * math.log(esr)) + (0.014 * pag)
            except ValueError: das28esr = 'nan'

            try: das28crp = (0.56 * math.sqrt(int(tjc))) + (0.28 * math.sqrt(sjc)) + (0.36 * math.log(crp +1)) + (0.014 * pag) + 0.96
            except ValueError: das28crp = 'nan'   
                          
            main_info_dict[patientID,'rf_tit'] = rf
            main_info_dict[patientID,'acpa_tit'] = acpa
            main_info_dict[patientID,'crp'] = crp
            main_info_dict[patientID,'esr'] = esr
            main_info_dict[patientID,'cdai'] = cdai
            main_info_dict[patientID,'sdai'] = sdai
            main_info_dict[patientID,'das28esr'] = das28esr
            main_info_dict[patientID,'das28crp'] = das28crp
            
            main_info_dict[patientID,'sjc28_base'] = sjc
            main_info_dict[patientID,'tjc28_base'] = tjc
            
            
            
print (count)
```

```python
#write output
output_file = '../../preprocessed_data/meta/patient_info_for_statistics.v2.tsv'
output_txt = open(output_file,'w')

variable_list = ['acpa', 'sex','age','bmi','rf','smoking','rf_tit','acpa_tit','crp','esr','cdai','sdai','das28esr','das28crp','rf_tit','acpa_tit','sjc28_base','tjc28_base']

output_txt.write('sample_ID\tmayo_ID')
for variable in variable_list:
    output_txt.write('\t%s' % variable)
output_txt.write('\n')

for patientID in patientID_list:
    mayo_ID = sample_mayoinfo_dict[patientID]
    output_txt.write('%s\t%s' % (patientID, mayo_ID))
                     
    for variable in variable_list:
        try: value = main_info_dict[patientID, variable]
        except KeyError: value = 'nan'
        output_txt.write('\t%s' % value)
    output_txt.write('\n')
output_txt.close()    
    
```
