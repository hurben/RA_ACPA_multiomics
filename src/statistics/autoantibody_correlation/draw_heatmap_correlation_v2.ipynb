```r
#correlation_heatmap   22.04.13
#draw heatmap of correlation - clinical covariates

library(gplots)
library(pheatmap)
library(RColorBrewer)
library(compositions)
```

```r
get_heatmap_results <- function(data_file, bin_file, defined_title){
    
    data_df <- read.csv(data_file, sep="\t", header=TRUE, row.names=1)
    data_df <- as.data.frame(data_df)
    
    bin_df <- read.csv(bin_file, sep="\t", header=TRUE, row.names=1)
    bin_df <- as.data.frame(bin_df)
    
    df_size <- dim(data_df)
    df_row_length <- df_size[1]
    df_col_length <- df_size[2]
    
    colorpal=colorRampPalette(c("#083254", "#FFFFFF", "#b35015"))(100)
    

    heatmap_results <- pheatmap(data_df, show_rownames=TRUE, show_colnames=TRUE, 
                                color = colorpal, fontsize_row = 5,
                                cluster_cols=FALSE, 
                                # cluster_rows=FALSE,
                                legend_labels=FALSE, 
                                treeheight_row=0,
                                main=defined_title, 
                                display_numbers = bin_df, cellheight=8, cellwidth=40)

    return (heatmap_results)
    
}
```

```r
data_file = '../../../analysis/statistics/autoantibody_correlation_v2/autoantibody_patient_info_correlation.acpa_neg.filtered.main.tsv'
bin_file = '../../../analysis/statistics/autoantibody_correlation_v2/autoantibody_patient_info_correlation.acpa_neg.filtered.bin.tsv'

heatmap_results <- get_heatmap_results(data_file, bin_file, "Correlation Heatmap\nAutoantibody-ClinicalCovariates (ACPA-neg RA)")

output_pdf_label = '../../../analysis/statistics/autoantibody_correlation_v2/autoantibody_patient_info_correlation.acpa_neg.filtered.pdf'
pdf(output_pdf_label, width=6, height=9)
print(heatmap_results)
dev.off()
```
