```python
#PREPROCESS_make_correlation_ready_profile_STEP3.ver_binary_V2     22.09.12
#This code by-pass a few logics from previous output (generated by PREPROCESS_make_correlation_ready_profile_STEP2)
#Caution is required when modifying this code: cell 2, Line28~34
#If previous output is modified, this script is recommended to be removed.

#update: 23.01.24
#considering acpa-neg, acpa-pos files


import pandas as pd
```

```python
def main(data_file, output_file):
    data_df = pd.read_csv(data_file, sep="\t", index_col = 0)
    r, c = data_df.shape

    alpha_feature_list = data_df.index.values
    beta_feature_list = data_df.columns.values

    output_txt = open(output_file,'w')
    #write header
    for beta_feature in beta_feature_list:
        output_txt.write('\t%s' % beta_feature)
    output_txt.write('\n')

    for i in range(r):
        alpha_feature = alpha_feature_list[i]
        sig_count = 0

        for beta_feature in beta_feature_list:
            corr = data_df[beta_feature][i]
            if corr != 0:
                if corr > 0.4 or corr < -0.4:
                    sig_count += 1

        if sig_count >= 1: #write feature and correlations 
            output_txt.write(alpha_feature)
            for beta_feature in beta_feature_list:
                corr = data_df[beta_feature][i]

                if corr != 0:
                    output_txt.write('\t*')
                else:
                    output_txt.write('\t.')
            output_txt.write('\n')

    output_txt.close()
```

```python
data_dir = '../../../analysis/statistics/autoantibody_correlation_v2'

data_file = '%s/autoantibody_patient_info_correlation.acpa_pos.tsv' % (data_dir)
output_file = '%s/autoantibody_patient_info_correlation.acpa_pos.filtered.bin.tsv' % (data_dir)
main(data_file, output_file)

data_file = '%s/autoantibody_patient_info_correlation.acpa_neg.tsv' % (data_dir)
output_file = '%s/autoantibody_patient_info_correlation.acpa_neg.filtered.bin.tsv' % (data_dir)
main(data_file, output_file)
```
