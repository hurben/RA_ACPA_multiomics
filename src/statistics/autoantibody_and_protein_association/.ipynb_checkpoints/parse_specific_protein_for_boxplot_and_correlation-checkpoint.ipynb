```python
import pandas as pd
import seaborn as sns
import matplotlib as mpl
from scipy.stats import spearmanr

mpl.rcParams['pdf.fonttype'] = 42
```

```python
result_df = pd.DataFrame(columns= ["Protein", "Autoantibody","ACPA"])
```

```python
proteomic_file = '../../../preprocessed_data/proteomics/somascan_anml.T.v2.tsv'
protein_df = pd.read_csv(proteomic_file, sep="\t", index_col=0)

patient_info_file = '../../../preprocessed_data/meta/patient_info.ML_ready.tsv'
patient_df = pd.read_csv(patient_info_file, sep="\t",index_col=0)

aa_file = '../../../preprocessed_data/autoantibody/sengenics_qnorm_data.v2.tsv'
aa_df = pd.read_csv(aa_file, sep="\t", index_col=0)

protein_df = protein_df.append(patient_df.loc["acpa"])
protein_df = protein_df.append(aa_df.loc["MUTYH"])
protein_df = protein_df.T
```

```python
subset_df = protein_df[["MUTYH_10030-8","MUTYH", "acpa"]]

subset_df.loc[subset_df["acpa"] == 1, "acpa"] = "ACPA-positive"
subset_df.loc[subset_df["acpa"] == 2, "acpa"] = "ACPA-negative"
subset_df.loc[subset_df["acpa"] == 0, "acpa"] = "Healthy Control"
```

```python
my_palette = {"ACPA-positive":"#F5918A", "Healthy Control":"#4FB95A", "ACPA-negative":"#8CB4E0"}

#AutoAntibody Box plot
ax = sns.swarmplot(x="acpa", y="MUTYH", data=subset_df, color=".35", size=4)
ax = sns.boxplot(x="acpa", y="MUTYH", data=subset_df, palette=my_palette)
ax.figure.savefig('../../../analysis/statistics/autoantibody_and_protein_association/MUTYH_aa_boxplot.pdf')
```

```python
#Protein Box plot
ax = sns.swarmplot(x="acpa", y="MUTYH_10030-8", data=subset_df, color=".35", size=4)
ax = sns.boxplot(x="acpa", y="MUTYH_10030-8", data=subset_df, palette=my_palette)
ax.set_ylim(50,500)
ax.figure.savefig('../../../analysis/statistics/autoantibody_and_protein_association/MUTYH_p_boxplot.pdf')
```
