```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

import matplotlib as mpl
mpl.rcParams['pdf.fonttype'] = 42

#Purpose: visualization
```

```python
data_file = "../../../preprocessed_data/meta/patient_info_for_statistics.tsv"
data_df = pd.read_csv(data_file, sep="\t")
```

```python
acpa_pos_df = data_df[data_df['acpa'] == 1]
acpa_neg_df = data_df[data_df['acpa'] == 2]
acpa_df = data_df[data_df['acpa'].isin([1, 2])]
```

```python
#DAS28-CRP
dist_plt = sns.displot(acpa_df, x="das28crp", hue="acpa", palette=["#F5918A", "#8CB4E0"], kde=True)
dist_plt.savefig('../../../analysis/visualization/Figure2_das28crp.pdf')
```

```python
dist_plt = sns.displot(data_df, x="age", hue="acpa", palette=["#4FB95A", "#F5918A", "#8CB4E0"], kde=True)
dist_plt.savefig('../../../analysis/visualization/Figure2_age.pdf')
```

```python
dist_plt = sns.displot(data_df, x="bmi", hue="acpa", palette=["#4FB95A", "#F5918A", "#8CB4E0"], kde=True)
dist_plt.savefig('../../../analysis/visualization/Figure2_bmi.pdf')
```
