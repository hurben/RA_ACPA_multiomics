```r
#DRAW_HEATMAP_patient_treatment
#
#Purpose: visualization
#Simply draw a heatmap that illustrates a heatmap for the patient treatment


library(gplots)
library(pheatmap)
library(RColorBrewer)
library(compositions)
```

```r
data_file = "../../../preprocessed_data/meta/patient_info_for_statistics.tsv"
data_df <- read.csv(data_file, sep="\t", header=TRUE, row.names=1)
data_df <- as.data.frame(data_df)
df_size <- dim(data_df)
df_row_length <- df_size[1]
df_col_length <- df_size[2]
```

```r
annotation <- data.frame(condition = factor(data_df[,2], labels = c("Control", "ACPA_positive", "ACPA_negative")))
rownames(annotation) <- rownames(data_df)
data_new_df <- data_df[,16:df_col_length]
```

```r
color_list = colorRampPalette(c("white","black"))(2)


heatmap_results <- pheatmap(data_new_df, show_rownames=FALSE, show_colnames=TRUE,
                            color = color_list, cluster_cols=TRUE, legend_labels=FALSE, cluster_rows=FALSE,
                           annotation_row = annotation)

output_pdf_label = '../../../analysis/visualization/patient_drug_info_heatmap.pdf'
pdf(output_pdf_label)
heatmap_results
dev.off()
```
