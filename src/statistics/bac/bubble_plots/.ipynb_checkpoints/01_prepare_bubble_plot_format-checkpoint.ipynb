```python
import pandas as pd
import math
```

```python
#For creating a bubble plot of:
#X-axis: ACPA-neg up, ACPA-neg down, ACPA-pos up, ACPA-pos down
#Y-axis: Pathways
#circle size: number of DAM/ pathway size
#circle color: -log10(p-value)

#make data matrix of
#1st column: Pathway name, 
#2nd column: condition
#3rd column: -log10(P-value)
#4th column: number of DAM / pathway size

pvalue_file = "../../../analysis/statistics/gse/metabolomics/metabolomics.cVSneg.down.hypergeometric.result.tsv"
pvalue_df = pd.read_csv(pvalue_file, sep="\t")
```

```python
data_dict = {}

pathway_list = []

condition_list = ["cVSneg", "cVSpos"]
regulation_list = ["up","down"]

for condition in condition_list:
    for regulation in regulation_list:
        label_pathway_list = []

        label = "%s_%s" % (condition, regulation)
                
        data_file = "../../../analysis/statistics/gse/metabolomics/metabolomics.%s.%s.hypergeometric.result.tsv" % (condition, regulation)
        data_df = pd.read_csv(data_file, sep = "\t")
        r, _ = data_df.shape
        
        for i in range(r):
            pathway = data_df["pathway"][i]
           
            pval = float(data_df["pvalue"][i])
            
            if pval < 0.05:
                logpval = -math.log10(pval)
                pathway_list.append(pathway)
                label_pathway_list.append(pathway)
            
                try: data_dict[label, pathway][0] = logpval
                except KeyError: data_dict[label, pathway] = [logpval, 'nan']

        data_file = "../../../analysis/statistics/gse/metabolomics/metabolomics.%s.%s.hypergeometric.ready.tsv" % (condition, regulation)
        data_df = pd.read_csv(data_file, sep = "\t")
        r, _ = data_df.shape
        
        for i in range(r):
            pathway = data_df["pathway"][i]
            
            if pathway in label_pathway_list:          
                denominator = data_df["number_of_metabolites_in_pathway(m)"][i]
                numerator = data_df["significant_metabolites_in_pathway(x)"][i]
                
                
                circle_size = numerator/denominator
                
                data_dict[label, pathway][1] = circle_size

pathway_list = list(set(pathway_list))
```

```python
output_file = "../../../analysis/statistics/gse/metabolomics/plot/bubble_plot_ready.tsv"
output_txt = open(output_file,'w')

output_txt.write("data_label\tpathway\tsignificance\tsize\n")

for condition in condition_list:
    for regulation in regulation_list:
        
        data_label = "%s_%s" % (condition, regulation)
        
        for pathway in pathway_list:
            
            try: logpval = data_dict[data_label, pathway][0]
            except KeyError: logpval = 0
            try: circle_size = data_dict[data_label, pathway][1]
            except KeyError: circle_size = 0
            print (data_label, pathway, logpval, circle_size)
            
            output_txt.write("%s\t%s\t%s\t%s\n" % (data_label, pathway, logpval, circle_size))
            
output_txt.close()
```
