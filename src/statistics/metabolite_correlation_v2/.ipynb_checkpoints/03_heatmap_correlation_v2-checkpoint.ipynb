```r
#correlation_heatmap   22.04.13
#draw heatmap of correlation - clinical covariates

library(gplots)
library(pheatmap)
library(RColorBrewer)
library(compositions)
```

```r
get_heatmap_results <- function(data_file, bin_file, defined_title){
    
    data_df <- read.csv(data_file, sep="\t", header=TRUE, row.names=1)
    data_df <- as.data.frame(data_df)
    
    bin_df <- read.csv(bin_file, sep="\t", header=TRUE, row.names=1)
    bin_df <- as.data.frame(bin_df)
    
    df_size <- dim(data_df)
    df_row_length <- df_size[1]
    df_col_length <- df_size[2]

    # colorpal <- brewer.pal(11,"RdBu")
    # colorpal <- c(colorpal[1:6], "#FFFFFF", colorpal[7:11])
    # colorpal <- rev(colorpal)
    
    # colorpal <- brewer.pal(11,"RdBu")
    # colorpal <- c(colorpal[1:6], "#FFFFFF", colorpal[7:11])
    # colorpal <- rev(colorpal)
    
    colorpal=colorRampPalette(c("#2676B6", "#FFFFFF", "#F5AE84"))(100)
    

    heatmap_results <- pheatmap(data_df, show_rownames=TRUE, show_colnames=TRUE, 
                                color = colorpal, fontsize_row = 3,
                                cluster_cols=TRUE, legend_labels=FALSE, 
                                main=defined_title, 
                                display_numbers = bin_df, cellheight=10)

    return (heatmap_results)
    
}
```

```r
data_file = '../../../analysis/statistics/metabolite_correlation_v2/metabolite_patient_info_correlation.acpa_neg.filtered.main.tsv'
bin_file = '../../../analysis/statistics/metabolite_correlation_v2/metabolite_patient_info_correlation.acpa_neg.filtered.bin.tsv'

heatmap_results <- get_heatmap_results(data_file, bin_file, "Correlation Heatmap\nMetabolites-ClinicalCovariates (ACPA-neg RA)")

output_pdf_label = '../../../analysis/statistics/metabolite_correlation_v2/metabolite_patient_info_correlation.acpa_neg.filtered.pdf'
pdf(output_pdf_label)
print(heatmap_results)
dev.off()
```
