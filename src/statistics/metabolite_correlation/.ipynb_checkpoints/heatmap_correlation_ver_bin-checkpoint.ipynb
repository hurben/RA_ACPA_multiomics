```r
#correlation_heatmap_ver_bin   22.09.12

#This script is a complementary heatmap to provide information for the other color heatmap.
#draw heatmap of correlation - clinical covariates

library(gplots)
library(pheatmap)
library(RColorBrewer)
library(compositions)
```

```r
get_heatmap_results <- function(data_file, defined_title){
    
    data_df <- read.csv(data_file, sep="\t", header=TRUE, row.names=1)
    data_df <- as.data.frame(data_df)
    
    df_size <- dim(data_df)
    df_row_length <- df_size[1]
    df_col_length <- df_size[2]

    heatmap_results <- pheatmap(data_df, show_rownames=TRUE, show_colnames=TRUE, 
                                color = c("white","black"), fontsize_row = 5,
                                cluster_cols=TRUE, legend_labels=FALSE, main=defined_title, 
                                display_numbers = round(data_df,0))

    return (heatmap_results)
    
}
```

```r
data_file = '../../../analysis/statistics/metabolite_correlation/metabolite_patient_info_correlation.filtered.v2.bin.tsv'

heatmap_results <- get_heatmap_results(data_file, "Correlation Heatmap\nMetabolites(n=53)-ClinicalCovariates (RA only)")

output_pdf_label = '../../../analysis/statistics/metabolite_correlation/metabolite_patient_info_correlation.filtered.v2.bin.pdf'
pdf(output_pdf_label)
heatmap_results
dev.off()
```
