```python
#03_organize_significant_features

#read from padj file

#get significant only (cohens' |d| > 0.5, P < 0.01, ignore autoantibodies)
#note: up/down regulated is not an issue

#these list will be further compared with:
#old popluation: upregulated in ACPA-neg RA, upregulated in ACPA-pos RA
#young popluation: upregulated in ACPA-neg RA, upregulated in ACPA-pos RA
```

```python
import pandas as pd

def find_truely_uniq(data_file, condition_of_interest, acpa_sig_list, control_sig_list):
    print ('---------------')
    data_open = open(data_file,'r')
    data_readlines = data_open.readlines()
    
    original_list = []
    filtered_list = []
    second_filtered_list = []
    
    print ("input acpa specific list: %s" % len(acpa_sig_list))
    print ("input control specific list: %s" % len(control_sig_list))
    
    #acpa filter (find common)
    for i in range(len(data_readlines)):
        read = data_readlines[i]
        read = read.replace('\n','')
        token = read.split('\t')
        
        condition = token[0]
        if condition == condition_of_interest:
            print (condition, condition_of_interest)
            for j in range(1, len(token)):
                feature = token[j]        
                if feature != "":
                    original_list.append(feature)                
                    if feature in acpa_sig_list:
                        filtered_list.append(feature)

    original_list = list(set(original_list))
    print ("%s/%s" % (len(filtered_list), len(original_list)))
    # print ("1st filter: same features: %s" % filtered_list)
    
    #control filter (should not be common)
    for feature in filtered_list:
        if feature not in control_sig_list:
            second_filtered_list.append(feature)

    original_list = list(set(original_list))
    print ("%s/%s" % (len(second_filtered_list), len(original_list)))
    print ("2nd filter: %s" % second_filtered_list)
    print ('---------------')    
    
                        
    data_open.close()
    
    return filtered_list

def find_truely_uniq_ver_protein(data_file, condition_of_interest, acpa_sig_list, control_sig_list):
    
    print ('---------------')
    data_open = open(data_file,'r')
    data_readlines = data_open.readlines()
    
    original_list = []
    filtered_list = []
    second_filtered_list = []
    
    print ("input acpa specific list: %s" % len(acpa_sig_list))
    print ("input control specific list: %s" % len(control_sig_list))
    
    #acpa filter (find common)
    for i in range(len(data_readlines)):
        read = data_readlines[i]
        read = read.replace('\n','')
        token = read.split('\t')
        
        condition = token[0]
        if condition == condition_of_interest:
            print (condition, condition_of_interest)
            for j in range(1, len(token)):
                feature = token[j]
                feature = 'p_%s' % feature
                if feature != "":
                    original_list.append(feature)                
                    if feature in acpa_sig_list:
                        filtered_list.append(feature)

    original_list = list(set(original_list))
    print ("%s/%s" % (len(filtered_list), len(original_list)))
    # print ("1st filter: same features: %s" % filtered_list)
    
    #control filter (should not be common)
    for feature in filtered_list:
        if feature not in control_sig_list:
            second_filtered_list.append(feature)

    original_list = list(set(original_list))
    print ("%s/%s" % (len(second_filtered_list), len(original_list)))
    print ("2nd filter: %s" % second_filtered_list)
    print ('---------------')    
    
    
    data_open.close()
    
    return filtered_list
    
def make_condition_specific_sig_list(data_name):

    data_dir = "../../../analysis/age_stratified/differential_abundance_acpa_specific"
    data_file = "%s/%s" % (data_dir, data_name)
    data_df = pd.read_csv(data_file, sep="\t", index_col=0)

    #get sig list
    r, c = data_df.shape
    feature_list = data_df.index.values

    sig_p_list = []
    sig_m_list = []
    data_dict = {}

    for i in range(r):
        feature = feature_list[i]
        cohen_d = data_df["cohenD"][i]
        pval = data_df["pval"][i]

        if pval < 0.01 and abs(cohen_d) > 0.5:
            data_dict[feature] = [cohen_d, pval]

            if 'p_' in feature:
                sig_p_list.append(feature)                
            if 'p_' not in feature and 'aa_' not in feature:
                sig_m_list.append(feature)
                
    return sig_p_list, sig_m_list, data_dict

def make_sig_list(acpa_pos_sig_list, acpa_neg_sig_list, control_sig_list, output_file):

    output_txt = open(output_file,'w')
    
    output_txt.write("acpa_pos_sig")
    for feature in acpa_pos_sig_list:
        output_txt.write('\t%s' % feature)
    output_txt.write('\n')
    
    output_txt.write("acpa_neg_sig")
    for feature in acpa_neg_sig_list:
        output_txt.write('\t%s' % feature)
    output_txt.write('\n')
    
    output_txt.write("control_sig")
    for feature in control_sig_list:
        output_txt.write('\t%s' % feature)
    output_txt.write('\n')

    output_txt.close()

#mostly for debugging.
```

```python
acpa_pos_p_list, acpa_pos_m_list, acpa_pos_data_dict = make_condition_specific_sig_list("high_low_acpa_pos.youngVSold.padj.tsv")
acpa_neg_p_list, acpa_neg_m_list, acpa_neg_data_dict = make_condition_specific_sig_list("high_low_acpa_neg.youngVSold.padj.tsv")
control_p_list, control_m_list, control_data_dict = make_condition_specific_sig_list("high_low_control.youngVSold.padj.tsv")

make_sig_list(acpa_pos_m_list, acpa_neg_m_list, control_m_list, "../../../analysis/age_stratified/differential_abundance_acpa_specific/condition_specific_sig_metabolites.tsv")
make_sig_list(acpa_pos_p_list, acpa_neg_p_list, control_p_list, "../../../analysis/age_stratified/differential_abundance_acpa_specific/condition_specific_sig_proteins.tsv")
```
